#### 4. To visualize data we will use matplotlib (even whe nwe use other libraries we are using matplotlib)

```python
x = npr.normal(0, .015, 100).cumsum()
x = 100 + 100 * x
```

```python
plt.plot(x)
```

```python
x = npr.standard_normal(30)
y = npr.random(30)
```

```python
plt.scatter(x,y, alpha=.5)
```

```python
z = npr.standard_normal(1000000)
```

```python
data = plt.hist(z, bins=50, color='g', edgecolor='w')
```

#### 5. The workhorse for us will be Pandas DataFrame

```python
a = npr.standard_normal((100, 5)).cumsum(axis=0)
```

```python
df = pd.DataFrame(100 + 100 * a, 
                 index= pd.date_range('20190101', freq='B', periods=100),
                 columns=list("ABCDE"))
```
